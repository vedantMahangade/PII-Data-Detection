```python
dataset
```

```python
# function to return flattened dataset for EDA
def flatten_dataset(dataset):
    flat_dataset = []
    textid = 0
    count = 0

    # iterate over each row in dataset
    for row in dataset:
        tokens = row['tokenised_unmasked_text']
        labels = row['token_entity_labels']

        # some rows have more labels than tokens hence ignoring those rows
        if len(tokens) == len(labels):
            count = count + 1

            # for each label create new row with corresponding token
            for i in range(len(tokens)):
                flat_dataset.append({ 'id':textid, 'token': tokens[i], 'labels': labels[i] })
        textid = textid + 1
    df = pd.DataFrame(flat_dataset, columns=['id', 'token','labels'])
    return df
```

```python
df = flatten_dataset(dataset)
df
```

```python
df = df[df['labels'] != 'O']

import matplotlib.pyplot as plt
frequency = df['labels'].value_counts()
plt.figure(figsize=(15, 8))

# Plot the frequency distribution
frequency.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Categorical Data')
plt.xticks(rotation=45)  # Rotate the x-axis labels by 45 degrees
plt.tight_layout() 
plt.show()
```
